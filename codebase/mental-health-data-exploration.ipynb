```python
import sqlite3

conn = sqlite3.connect('../data/mental_health.sqlite')
print("Connexion à la base de données ouverte avec succès")
```

```python
# Déterminons maintenant quelles sont les tables disponibles dans cette base
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```
